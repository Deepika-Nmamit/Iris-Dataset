```python
def plot_incorrect_predictions(df_pred, x_axis_feature, y_axis_feature):
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))
    axs = axs.flatten()
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="prediction_label", data=df_pred, ax=axs[0])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="target_name", data=df_pred, ax=axs[1])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="predict_correct", data=df_pred, ax=axs[2])
    axs[3].set_visible(False)
    plt.show()
```

```python
plot_incorrect_predictions (df_pred, "petal length (cm)", "petal width (cm)" )
```
